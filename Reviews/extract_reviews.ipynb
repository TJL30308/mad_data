```python
# Dependencies
import pandas as pd
import pycountry
```

## Load CSV file

```python
reviews = pd.read_csv("resources/winemag-data-130k-v2.csv")
reviews.head()
```

```python
reviews.rename(columns={'Unnamed: 0': "review_id"}, inplace=True)
reviews.head()
```

```python
reviews.set_index('country', inplace=True)
```

```python
reviews.head()
```

```python
# Create lists for country name and codes
name_list = []
country_codes = []
for country in pycountry.countries:
    name_list.append(country.name)
    country_codes.append(country.numeric)
```

```python
# Rename values
reviews.rename(index={
    "Czech Republic": "Czechia",
    "England": "United Kingdom",
    "Macedonia": "North Macedonia",
    "Moldova": "Moldova, Republic of",
    "US": "United States"
}, inplace=True)

reviews
```

```python
# Create dict for countries
country_dict = dict(zip(name_list, country_codes))
```

```python
# Create column to add to DF
new_col=[]

# Loop through indexes and get country_id
for row, value in reviews.iterrows():
    if row in country_dict:
        new_col.append(country_dict[row])

    else:
        new_col.append('000')

reviews['country_id'] = new_col
reviews.reset_index(inplace=True)
```
